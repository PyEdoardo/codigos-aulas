# Ciclo de Vida do Software

- Foi utilizado uma explicação fazendo um paralelo entre seres vivos e softwares, em que os seres vivos tem etapas como o nascimento, o crescimento, reproduzir-se e morrer.

- O Nascimento seria a criação e todos os estágios iniciais do software, o crescimento seria o de implementação de novas funcionalidades, a reprodução seria 

1° O Ciclo de Vida do Software é o termo utilizado para definir um conjunto de etapas que ocorre na concepção de um sistema e o instante em que ele é entregue pelo dev, ele ajuda a orientar a equipe e os gestores, e no direcionamento de recursos.


## Etapas do Ciclo de Vida do Software

- 1°: **Engenharia de Sistemas:**

    - O Seu foco encontra-se em definir de maneira precoce, as necessidades do cliente e suas funcionalidades.

    - Para a escolha de alternativa viável do software que será desenvolvido, se faz o estudo de viabilidade técnica, financeira, legal e etc.

    ° *Exemplo de Solução para estoque:*

        "Para ter um controle de estoque preciso de um software para cadastrar todos os produtos atribuindo a informação de estoque (mínimo e máximo), gerador de inventário, fornecedores, compras automatizadas..."

        ° Perguntas pertinentes ao domínio:

            - Como devem ser organizados os produtos?

            - São quantos fornecedores?

            - Existem produtos sem identificação?

            - Há funcionários para registrar todas as entradas e saídas?

        ° Considerações para escolha da alternativa mais viável:

            - O Software deve ser construído dentro de qual prazo?

            - Existem tecnologias exigidas para o desenvolvimento?

            - Os componentes (hardware, peças) estão disponíveis?

            - Existe pessoal treinado?

- 2°: **Análise de Requisitos:**

    ° *Processo de descoberta, refinamento, modelagem e especificação dos requisitos de software.*

        - O Escopo do software identificado na etapa anterior é refinado.

    ° *Principal Atividade: Os Requisitos do sistema devem ser obtidos atráves de consultas aos stakeholders.*

        - Ou também atráves de documentações, outros sistemas, conhecimento do domínio e estudos de mercado.

    -> *Este processo é também conhecido como aquisição de requisitos ou descoberta de requisitos.*

        - Os requisitos são analisados em detalhe e os diferentes stakeholders negociam para decidirem que requisitos serão aceitos;

            - São utilizados como base para o desenvolvmento do software.

        - Atividade da Análise de Requisitos:

            - Reconhecimento detalhado do problema;

            - Definição da Solução do problema;

            - Modelagem (diagramas);

            - Revisão;

    ° *Habilidades do(a) Analista de Requisitos:*

        - Compreender conceitos abstratos e reorganizá-los de forma lógica;

        - Absorver informações pertinentes de fontes conflitantes e confusas (clientes/usuários);

        - Entender o ambiente do usuário e cliente

        - Desenvolver boa comunicação verbal.

    ° *Princípios da análise:*

        - O Domínio de um problema deve ser representado e compreendido;

        - O Motivo da criação de cada requisito deve ser lembrado;

        - O Problema pode/deve ser dividido em partes;

        - O Processo de análise deve mover-se da informação essencial para um nível mais detalhado (dos modelos de alto nível para os de baixo nível);

        - Os requisitos devem ser priorizados pelos usuários;

- 3°: Projeto de Software:

    ° Cada um dos artefatos gerados nas etapas anteriores fornecem subsídios para a criação de um modelo de projeto de software.

    ° O projeto deve representar todos os requisitos explicitos contidos na análise de requisitos e os requisitos implícitos desejados pelo cliente e/ou necessários para o sistema.

    ° O projeto deve ser **legível** pelos implementadores, testadores e demais integrantes do time.

        - Baseado nos requisitos analisados deve-se: planejar cronograma, fazer o design da interface (prototipação), design da arquitetura e etc;
